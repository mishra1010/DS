```python
#Read data from excel and use converters to convert data to meaningful info in columnsusing converters
#convert messy data in to writing something meaningful
import pandas as pd

def convert_people_cell(cell):
    if cell=="n.a":
        return 'sam'
    return cell
def convert_EPS_cell(cell):
    if cell=="not available":
        return None
    return cell
df = pd.read_excel("C:\\Data Science\\Pandas\\stockdata.xlsx","Sheet1",converters ={
        'People': convert_people_cell,
        'Eps': convert_EPS_cell
})
df
```

```python
#write to excel file
df.to_excel("new.xlsx",sheet_name="stocks")
```

```python
##write to excel file with offset and no index
df.to_excel("new.xlsx",sheet_name="stocks", startrow=1, startcol=2, index=False)
```
