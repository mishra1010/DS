```python
import pandas as pd
import numpy as np
df = pd.read_csv("C:\\Users\\demishra\\Documents\\DS\\weather_datawithnumsinmissingvalues.csv")
df
```

```python
#replace all values with 9999 as NAN
new_df = df.replace(-9999,np.NaN)
new_df
```

```python
#what if we have 2 special num values
#new_df = df.replace([-9999,-8888],np.NaN)
```

```python
#what if we want to replace different column with different values
new_df = df.replace({
    'Temperature':-9999,
    'windspeed':-9999,
    'Event':'0'
},np.NaN)
new_df
```

```python
#Also we can replace values differently by giving only nums to be replaced
new_df = df.replace({
    '9999: np.NaN,
    'No Event':'Sunny'
},np.NaN)
new_df
#replace using regex
new_df = df.replace({
    'Temperature': '[A-Za-z]'',
    'windspeed': '[A-Za-z]' 
},'',regex=True)
```
